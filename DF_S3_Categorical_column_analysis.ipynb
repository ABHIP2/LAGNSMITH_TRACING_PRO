```python
## step1   Create a folder
folder_name='Frequency_table'
os.makedirs(folder_name, exist_ok=True)

## step 2   make a new path
directory=os.getcwd()
new_dir=os.path.join(directory,folder_name)

## step3 : Create A DataFrame
for i in cat[1:]:
    keys=visadf[i].value_counts().keys()
    values=visadf[i].value_counts().values
    df=pd.DataFrame(zip(keys,values),
               columns=['Classes','Frequency'])

    ## step 4 make a new location 
    file_name=f'{i}_df.csv'
    path=os.path.join(new_dir,file_name)
    
    ## step5  save into DataFrame
    df.to_csv(path,index=False)
```

```python
for i in cat[1:]:
    keys=visadf[i].value_counts().keys()
    values=visadf[i].value_counts().values
    plt.figure(figsize=(10,4))
    plt.pie(values,
        labels=keys,
       autopct='%1.2f%%',
       startangle=45)
    plt.xlabel('Classes')
    plt.ylabel('Frequency')
    plt.title(f'{i}_Piechart')
plt.show()
```

```python
folder_name='Frequency_table'
os.makedirs(folder_name, exist_ok=True)

## step 2   make a new path
directory=os.getcwd()
new_dir=os.path.join(directory,folder_name)

for i in cat[1:]:
    keys=visadf[i].value_counts().keys()
    values=visadf[i].value_counts().values
    plt.figure(figsize=(10,4))
    plt.bar(x=keys,height=values)
    plt.xlabel('Classes')
    plt.ylabel('Frequency')
    plt.title(f'{i}_Pie Chart')
    plt.show()

    ## make a new path
    filename=f'{i}.jpg'
    path=os.path.join(new_dir,filename)

    ## save into DataFrame
    plt.savefig(path)
```

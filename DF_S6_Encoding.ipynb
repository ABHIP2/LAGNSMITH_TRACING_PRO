## Encoding

- Encoding means catogerical data to numerical data

- ML/DL or any models works based on maths

- So models cannot understanmd english charactors 

- So it is very important convert categorical label to numerical values

- For example Gender column has two lables

    - Male

    - Female

- Then Female represent as 0 , Male represent as 1 based on alphabatical order

```python
visadf['case_status'].unique()
```

- Certified should assign as 0

- Denied should assign as 1

```python
## Mapping

d={'Cerified':0,'Denied':1}
d
```

```python
visadf['case_status']=visadf['case_status'].map(d)
```

    ## OneHotEncoder
    
    - First how many lables are there in a column

    - For example case status has two examples
    
        - Certified
    
        - Denied

    - At a time one column value beacome 1 then other column value become 0

    |case_status|case_status_Certified|case_status_Denied|
    |-|
    |Certified|
    |Denied|

|case_status|case_status_Certified|case_status_Denied|

|Denied||0||1|

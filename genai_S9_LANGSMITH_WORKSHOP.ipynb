```python
## IN that the tokens == No of words in the ans you have


from langchain_groq import ChatGroq
from google.colab import userdata
API_KEY=userdata.get('GEMINI2_AP_KEY')

llm = ChatGroq(
    api_key=API_KEY,
    model='llama-3.3-70b-versatile',
    temperature=0,
    max_tokens=100,
    # reasoning_format="parsed",
    # timeout=None,
    # max_retries=2,
    # other params...
)
```

```python
## the answer is coming in unuseual format

llm.invoke('Tell me About Maharastra')
```

```python
## When the answer is coming in unuseual format
## Use the .content

llm.invoke('Tell me About Maharastra').content
```

```python
## import StrOutputParser
from langchain_core.output_parsers import StrOutputParser
## it clean our output answer


from langchain_groq import ChatGroq

from google.colab import userdata
API_KEY=userdata.get('GEMINI2_AP_KEY')
llm = ChatGroq(
    api_key=API_KEY,
    model='llama-3.3-70b-versatile',
    temperature=0,
    max_tokens=None,
)


parser=StrOutputParser()

chain=llm|parser
response=chain.invoke('Tell me about Indian Army')
print(response)
```

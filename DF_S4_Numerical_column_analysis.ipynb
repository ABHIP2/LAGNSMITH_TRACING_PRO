## Empirical Rule

```python
# u-1*sigma  to u+1*sigma  = 68
# step1 : means of wage
# step2 : std  of wage
# step3 : lb = mean-1*sigma
#         lb = s1 - 1* s2
# step4 : ub = mean +1*sigma
```

```python
wage_mean=visadf['prevailing_wage'].mean()
wage_std=visadf['prevailing_wage'].std()
lb=wage_mean-1*wage_std
ub=wage_mean+1*wage_std
lb,ub
```

```python

wage_data=visadf['prevailing_wage']
wage_mean=visadf['prevailing_wage'].mean()
wage_std=visadf['prevailing_wage'].std()
lb=wage_mean-1*wage_std
ub=wage_mean+1*wage_std
lb,ub
con1=wage_data>lb
con2=wage_data<ub
cond=con1 & con2
len(visadf[cond])
```

```python
len(visadf[cond])==68*25480/100
```
